```python
from decision_tree import DecisionTree

dataset = [
    ['sunny',85,85,'false','no'],
    ['sunny',80,90,'true','no'],
    ['overcast',83,86,'false','yes'],
    ['rainy',70,96,'false','yes'],
    ['rainy',68,80,'false','yes'],
    ['rainy',65,70,'true','no'],
    ['overcast',64,65,'true','yes'],
    ['sunny',72,95,'false','no'],
    ['sunny',69,70,'false','yes'],
    ['rainy',75,80,'false','yes'],
    ['sunny',75,70,'true','yes'],
    ['overcast',72,90,'true','yes'],
    ['overcast',81,75,'false','yes'],
    ['rainy',71,91,'true','no']
]
```

```python
DecisionTree(verbose=True).train(dataset, DecisionTree.information_gain).infer(['overcast',60,62,'false'])
```

```python
DecisionTree(verbose=True).train(dataset, DecisionTree.gain_ratio).infer(['overcast',60,62,'false'])
```

```python
DecisionTree(verbose=True).train(dataset, DecisionTree.gini_index).infer(['overcast',60,62,'false'])
```
